Sistema de bits: Cada byte, que representa um píxel, é formado por 8 bits.


[0,0,0,1,0,1,0,0]

Bit 7 é o mais significativo, pois implica em mais informações, enquanto o 0 é o menos significativo.

```python
import time
```

```python
def calc_pixel(npi,img2):
    img = img2
    c1 = 0
    c2 = 0
    for i in img:
        for j in i:
            p = j
            bits = [0]*8
            for k in range(8):
                if p < 2:
                    bits[k] = p
                    break
                elif p%2 == 0:
                    bits[k] = 0
                else:
                    bits[k] = 1
                p = p//2
            if bits[npi] == 0:
                img[c1][c2] = 0
            else:
                img[c1][c2] = 2**(npi)
            c2+=1
        c2 = 0
        c1 += 1
    return img
```

```python
ts = time.time()
img2 = calc_pixel(1,cv2.imread(path+"Fig0314(a)(100-dollars).tif",cv2.COLOR_BGR2GRAY))
tr = time.time()
print(tr-ts)
mostrar_img(imagem_dolar1,"Dolar americano","cinza")
mostrar_img(calc_pixel(7,cv2.imread(path+"Fig0314(a)(100-dollars).tif",cv2.COLOR_BGR2GRAY)),
            "Dolar americano- bit mais influente","cinza2")
mostrar_img(calc_pixel(0,cv2.imread(path+"Fig0314(a)(100-dollars).tif",cv2.COLOR_BGR2GRAY)),
            "Dolar americano- bit menos influente","cinza2")
```
