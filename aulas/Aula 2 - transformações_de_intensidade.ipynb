```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
from PIL import Image
```

```python
def mostrar_img(img, title, color = None):
    plt.figure(figsize = (5,5))
    plt.title(title)
    if color == "cinza":
        plt.imshow(img, cmap = "gray",vmin=0, vmax=255)
    else:
        plt.imshow(img)
    plt.show()
```

```python
path = "C:\\Users\\mateus\\Desktop\\processamento_de_imagens\\Aula_2\\imagens\\"
```

```python
imagem1 = cv2.imread(path+"424.jpg")
imagem1 = imagem1[:,:,::-1]
mostrar_img(imagem1,"Imagem colorida","cinza")
type(imagem1)
```

```python
imagem1 = cv2.cvtColor(imagem1, cv2.COLOR_RGB2GRAY)
mostrar_img(imagem1,"Imagem em escala de cinza","cinza")
type(imagem1)
```

Filtragem espacial: Cada pixel é independente dos seus vizinhos.

## Funções básicas: Negativo da imagem

Sendo a imagem uma função f(x,y), a negativa será -f(x,y). Ou seja, quanto mais claro,

mais escuro será o pixel na imagem negativa.

```python
#perguntar, tem diferença de -imagem1 para 255 - imagem1?
imagem1_negativa = 255-imagem1
mostrar_img(imagem1_negativa,"Imagem negativa","cinza")
```

## Funções básicas: Transformação logarítmica

Dado a imagem como função f(x,y), temos que a fl(x,y) = log(f(x,y)+1)

Função útil para modificação do contraste. Multiplicar por números maiores que 1 aumentam

o contraste e por números menores que 1 diminuem o contraste.

```python
#preparação da imagem para o padrão de escala entre 0 a 1 (dividir por 255)
imagem_1 = np.double(imagem1/255)
#soma-se 1, para ficar de 1 a 2
imagem_1 = imagem_1+1
#aplica-se o logarítimo na imagem
imagem_1 = np.log(imagem_1)
#retorna-se para a escala de 255 e pode-se multiplicar ou dividir para clarear e escurecer a imagem
imagem_1 = imagem_1*255*3

mostrar_img(imagem_1,"Imagem em cinza com transformação logarítmica","cinza")
```

## Funções de potências

Nesse método, fp(x,y) = f(x,y)^a. Serve para correção de gama. Valores de a maiores do que 1 aumentam o contraste e menores tem diminuição de contraste.

```python
imagem_p = imagem1/255
imagem_p = imagem_p**(2.5)
imagem_p = imagem_p*255
mostrar_img(imagem_p,"Imagem em cinza com transformação logarítmica","cinza")
```
