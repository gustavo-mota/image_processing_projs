Exemplo 2: Imagem simples formada por uma matriz

```python
def mostrar_img(img, title,color = None):
    plt.figure(figsize = (5,5))
    plt.title(title)
    if color == "cinza":
        plt.imshow(img, cmap = "gray")
    else:
        plt.imshow(img)
    plt.show()
```

```python
img = np.array([[1,0,0],[0,1,0],[0,0,1]])
print(img)
mostrar_img(img, "Matriz em escala de cinza","cinza")
```

Escala de cinza:


Escala formada por 1 byte (256 probabilidades de 0 a 255) que informa
a placa de video a escala de intensidade
de luz que será exibida na tela. Imgshow aproxima e relaciona qualquer matriz a essa escala.

```python
img = np.array([[122.5,0,0],[0,255,0],[0,0,100]])
print(img)
mostrar_img(img, "Matriz em escala de cinza","cinza")
```

```python
img = np.uint8([[255,0,0],[0,12,0],[0,0,50.5]])
print(img)
mostrar_img(img, "Matriz em escala de cinza","cinza")
```

```python
path2 = "C:\\Users\\mateus\\Desktop\\processamento_de_imagens\\Aula_1\\imagens\\"
imagem2 = cv2.imread(path2+"424.jpg")

imagem2 = imagem2[:,:,::-1] #correção do padrão BGR para RGB
imagem2_vermelho = imagem2[:,:,0:1]
imagem2_verde = imagem2[:,:,1:2]
imagem2_azul = imagem2[:,:,2:]

mostrar_img(imagem2, "Imagem 424: Pavimento com danos","cinza")
mostrar_img(imagem2_vermelho, "Imagem 424: Pavimento com danos-valor vermelho","cinza")
mostrar_img(imagem2_verde, "Imagem 424: Pavimento com danos-valor verde","cinza")
mostrar_img(imagem2_azul, "Imagem 424: Pavimento com danos-valor azul","cinza")
```
