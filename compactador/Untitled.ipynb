### Histograma da imagem

```python
def histograma(img, num_color):
    Dist_color = [0]*256
    for i in range(len(img)):
        for j in range(len(img[0])):
            Dist_color[img[i][j][num_color]] += 1
    
    return (Dist_color)
print(histograma(img, 0))
```

```python
img_binario = ""
start = time.time()
for i in range(img.shape[0]):
    for j in range(img.shape[1]):
        for k in range(img.shape[2]):
            img_binario+= '{0:08b}'.format(img[i][j][k])
print(len(img_binario))
print(len(img_binario)/8)
end = time.time()

print("tempo gasto: %d" %(end - start))
```

### Código de Colomb-Rice

```python
def compact_cr(img, number = 2):
    img_compact = ""
    for i in range(img.shape[0]):
        for j in range(img.shape[1]):
            for k in range(img.shape[2]):
                
                img_int = '{0:b}'.format(img[i][j][k]//number)
                img_rest = '{0:b}'.format(img[i][j][k]%number)
                img_compact+= img_int
                img_compact+= img_rest
    return img_compact

start = time.time()

img_compactada = compact_cr(img, 4)

end = time.time()

print("tempo gasto: %d" %(end - start))

print(len(img_compactada))
print(len(img_compactada)/8)
print(len(img_compactada)/len(img_binario))
```
